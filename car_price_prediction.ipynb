# Feature Selection

```python
x1=X.iloc[:,:-1]
y=X.iloc[:,-1]
```

```python
x1
```

```python
y
```

# Data Splitting

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x1,y,test_size=0.2, random_state=9897)
```

# Linear Regression Model

```python
l_model=LinearRegression()  #object
l_model.fit(x_train,y_train) #function 
```

```python
prediction=l_model.predict(x_test)
prediction
```

# Model Evaluation

```python
from sklearn.metrics import *
```

```python
print("Mean absolute error: ",mean_absolute_error(y_test,prediction))
print("Root Mean squared error: ",mean_squared_error(prediction,y_test,squared=False))
print("Mean absolute percentage error: ",mean_absolute_percentage_error(y_test,prediction))
print("Accuracy score: ",r2_score(y_test,prediction))
```

```python
train_predict=l_model.predict(x_train)
```

```python
print("Mean absolute error: ",mean_absolute_error(y_train,train_predict))
print("Root Mean squared error: ",mean_squared_error(y_train,train_predict,squared=False))
print("Mean absolute percentage error: ",mean_absolute_percentage_error(y_train,train_predict))
print("Accuracy score: ",r2_score(y_train,train_predict))
```

# Cross-Validation

```python

from sklearn.model_selection import cross_val_predict, cross_val_score

# Create a Linear Regression model
l_model = LinearRegression()

# Perform 5-fold cross-validation
cv_predictions = cross_val_predict(l_model, x1, y, cv=5)

# Calculate cross-validation scores
cv_scores = cross_val_score(l_model, x1, y, cv=5)

# Evaluate the model
print("Cross-Validation Mean absolute error: ", mean_absolute_error(y, cv_predictions))
print("Cross-Validation Root Mean squared error: ", mean_squared_error(cv_predictions, y, squared=False))
print("Cross-Validation Mean absolute percentage error: ", mean_absolute_percentage_error(y, cv_predictions))
print("Cross-Validation R-squared (Accuracy score): ", r2_score(y, cv_predictions))

# Print cross-validation scores for each fold
print("Cross-Validation Scores:", cv_scores)
```

# Hyperparameter tuning

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x1_scaled = scaler.fit_transform(x1)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Define a range of alpha values to search
param_grid = {'alpha': [0.1, 1, 10, 100]}

# Create a GridSearchCV object for Ridge regression
grid_search = GridSearchCV(Ridge(), param_grid, cv=5, scoring='neg_mean_squared_error')

# Perform the search
grid_search.fit(x1_scaled, y)

# Get the best alpha value
best_alpha = grid_search.best_params_['alpha']
print("Best Alpha:", best_alpha)

# Create a Ridge regression model with the best alpha value
final_model = Ridge(alpha=best_alpha)
final_model.fit(x_train, y_train)

# Make predictions on the test set with the tuned model
tuned_predictions = final_model.predict(x_test)


```

```python
#predictions on the test set with the tuned model
tuned_predictions = final_model.predict(x_test)

# Evaluate the tuned model
print("Mean absolute error (tuned model): ", mean_absolute_error(y_test, tuned_predictions))
print("Root Mean squared error (tuned model): ", mean_squared_error(tuned_predictions, y_test, squared=False))
print("Mean absolute percentage error (tuned model): ", mean_absolute_percentage_error(y_test, tuned_predictions))
print("Accuracy score (tuned model): ", r2_score(y_test, tuned_predictions))
```
